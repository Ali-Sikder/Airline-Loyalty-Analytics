```python
import pandas as pd

customer_flight = pd.read_excel("Customer Flight Activity.xlsx")
customer_loyalty = pd.read_excel("Customer Loyalty History.xlsx")
calendar = pd.read_excel("Calendar.xlsx")
```

```python
customer_loyalty["Salary"] = customer_loyalty["Salary"].astype("Int64")
customer_loyalty.dtypes
```

```python
customer_loyalty["Customer Status"] = customer_loyalty["Cancellation Year"].apply(
    lambda x: "Active" if pd.isna(x) else "Cancelled"
)
```

```python
calendar["Year"] = calendar["Date"].dt.year
calendar["Month"] = calendar["Date"].dt.month
calendar.head()
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("airline_loyalty.db")
```

```python
customer_flight.to_sql("customer_flight", conn, if_exists="replace", index=False)
customer_loyalty.to_sql("customer_loyalty", conn, if_exists="replace", index=False)
calendar.to_sql("calendar", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
pd.read_sql("""
SELECT "Loyalty Number",
       SUM("Total Flights") AS Total_Flights
FROM customer_flight
GROUP BY "Loyalty Number"
ORDER BY Total_Flights DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql("""
SELECT "Customer Status",
       COUNT(*) AS Total_Customers
FROM customer_loyalty
GROUP BY "Customer Status";
""", conn)
```

```python
pd.read_sql("""
SELECT Year,
       SUM("Total Flights") AS Total_Flights
FROM customer_flight
GROUP BY Year
ORDER BY Year;
""", conn)
```

```python
pd.read_sql("""
SELECT "Loyalty Number",
       SUM(Distance) AS Total_Distance
FROM customer_flight
GROUP BY "Loyalty Number"
ORDER BY Total_Distance DESC
LIMIT 10;
""", conn)
```
